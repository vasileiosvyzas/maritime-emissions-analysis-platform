# Compare dataframes if they have the same schema

```python
def compare_dataframe_schemas(df1, df2):
    # Get the column names and dtypes for both DataFrames
    schema1 = pd.DataFrame({'column': df1.columns, 'dtype': df1.dtypes})
    schema2 = pd.DataFrame({'column': df2.columns, 'dtype': df2.dtypes})

    # Sort both schemas by column name for consistent comparison
    schema1 = schema1.sort_values('column').reset_index(drop=True)
    schema2 = schema2.sort_values('column').reset_index(drop=True)

    # Compare the schemas
    if schema1.equals(schema2):
        print("The schemas are identical.")
    else:
        print("The schemas are different.")
        
        # Find columns in df1 that are not in df2
        missing_in_df2 = set(df1.columns) - set(df2.columns)
        if missing_in_df2:
            print(f"Columns in df1 but not in df2: {missing_in_df2}")
        
        # Find columns in df2 that are not in df1
        missing_in_df1 = set(df2.columns) - set(df1.columns)
        if missing_in_df1:
            print(f"Columns in df2 but not in df1: {missing_in_df1}")
        
        # Compare dtypes for common columns
        common_columns = set(df1.columns) & set(df2.columns)
        for col in common_columns:
            if df1[col].dtype != df2[col].dtype:
                print(f"Column '{col}' has different dtypes: {df1[col].dtype} in df1, {df2[col].dtype} in df2")
```

```python
compare_dataframe_schemas(df_2023v12, df_2022v219)
```

```python
def add_generation_date(df, date):
    df['Generation_date'] = date
    df['Generation_date'] = pd.to_datetime(df['Generation_date'], format='%d%m%Y')
    
    return df
```
