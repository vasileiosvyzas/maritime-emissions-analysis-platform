```python
import pandas as pd
import requests
import time

from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('../data/raw/sflcorp_fleet.csv')
```

```python
df.head()
```

```python
wrong_page_names = []
ship_info = {}
for index, df_row in df.iterrows():
    print(df_row['Vessel'])
    vessel_name = df_row['Vessel'].lower().replace(' ', '-')
    print(vessel_name)
    
    html_page = requests.get(f"https://www.sflcorp.com/fleet/{vessel_name}")
    soup = BeautifulSoup(html_page.text, "html.parser")
    page_text = soup.text
    if 'Page not found' not in page_text:
        data = {}

        # Find all rows
        rows = soup.find_all('div', attrs={'class': 'flex-nowrap'})
        for row in rows:
            # Extract the label and value
            label = row.find('span', class_='label').text.strip().replace(':', '')
            value = row.find_all('span')[1].text.strip()
            # Add to dictionary
            data[label] = value
        ship_info[df_row['Vessel']] = data
    else:
        wrong_page_names.append(vessel_name)
        
    time.sleep(5)
    
```
